## Desenvolvimento

* Importação de todas bibliotecas necessárias para o desenvolvimento

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
import numpy as np
import matplotlib.pyplot as plt
```

* Divisão do dataset entre features e target

```python
X = dataset.data
Y = dataset.target
```

* Divisão do dataset para treinamento e teste, foi utilizado 25% do dataset para teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, shuffle=True)
```

###  Redes Neurais - Classificação por Multi Layer Perceptron 

- Será utilizado o modelo MLP para classificação dos cultivadores de vinho, utilizando 30 neurônios e posteriormente verificado com o aumento dos mesmos

```python
from sklearn.neural_network import MLPClassifier
```

```python
ANN = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(10))
```

```python
ANN.fit(X_train, y_train)
```

- Avaliando o desempenho do classificador por meio das métricas de acurácia e matriz de confusão

```python
y_pred_ANN = ANN.predict(X_test)

accuracy_score(y_test, y_pred_ANN)
```

```python
confusion_matrix(y_test, y_pred_ANN)
```

- Será feita uma análise, iterando o número de neurônios entre 30 e 500, afim de obter sua influência na acurácia da classificaçãom

```python
accuracy_values=[]

for i in range(30,50):
    ANN = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(i))
    
    
    ANN.fit(X_train, y_train)
    
    y_pred_ANN = ANN.predict(X_test)
    
    accuracy_values.append(accuracy_score(y_test, y_pred_ANN))
```

- Verificando graficamente a correlação entre o número de neurônios e a acurácia

```python
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

params = {'figure.figsize': [10, 6], 
          'axes.labelsize': 16,
          'axes.titlesize':18, 
          'font.size': 16,
          'legend.fontsize': 12, 
          'xtick.labelsize': 12, 
          'ytick.labelsize': 12
         }

plt.rcParams.update(params)

# Accuracy graph
for value in accuracy_values:
    plt.scatter(accuracy_values.index(value), value)
plt.legend()
plt.title('Acurácia do modelo de redes neurais')
plt.xlabel('Neuronios')
plt.ylabel('Valor')
```

### Clusterização por KNN

```python
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

params = {'figure.figsize': [10, 6], 
          'axes.labelsize': 16,
          'axes.titlesize':18, 
          'font.size': 16,
          'legend.fontsize': 12, 
          'xtick.labelsize': 12, 
          'ytick.labelsize': 12
         }

plt.rcParams.update(params)

# Accuracy graph
for value in accuracy_values_knn:
    plt.scatter(accuracy_values_knn.index(value), value)
plt.legend()
plt.title('Acurácia do modelo KNN')
plt.xlabel('Vizinhos')
plt.ylabel('Valor')
```

Já a árvore de decisões mostrou-se com um bom desempenho, embora ainda não o ideal, para classificar os vinhos conforme seu cultivador

Pela matriz de confusão torna-se bem claro que o modelo torna-se enviezado para apenas uma classe de vinho, enquanto os falsos positivos aparecem nas outras 2 classes restantes.
Já a arvore de decisão dá proveito das decisões de features marcantes que evidenciam apenas pela sua fronteira de valor que o vinho se refere a determinada classe, não necessitando de um dataset tão grande para atingir uma acurácia com valor razoável
